### 1. Basic of Matplotlib

````python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline 

````

#### A Simple Starter plot

````python
x_data = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y_data = np.exp(x_data) # e^(x)
````

````python
plt.plot(x_data, y_data) # Parameter 1 is an array containing the points on the x-axis.
                         # Parameter 2 is an array containing the points on the y-axis.

plt.title("This is a title")
plt.xlabel('X_data, normally axis Title')
plt.ylabel('Y_data, noramlly Y axis Title')
````

### Subplots

`plt.subplot` function you can draw multiple plots in one figure

````python
# plt.subplot(number of rows, number of columns, plot number)

x = np.array([1, 2, 3, 4])
y = np.array([1, 4, 9, 16])

# first plot
plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.
plt.plot(x, y)




x = np.array([1, 2, 3, 4])
y = np.array([10, 6, 50, 40])

# second plot 
plt.subplot(1, 2, 2)
plt.plot(x, y) 

````

### Figure Size

You can use `plt.figure(figsize=(width, height))` to adjust the figure size to your desied size

````python
plt.figure(figsize=(10, 8))

plt.plot(x_data, y_data)
````

You can also specify the dot per inch (DPI)

````python
plt.figure(figsize=(4,3), dpi=100)

plt.plot(x_data, y_data)
````

`figsize=(width, height)` can also be used in subplots that we saw earlier.



```
plt.subplots(figsize=(10,5))

```

#### Figure Title, Axis Labels and Legends

The `linspace()` method creates an array with evenly spaced elements over an interval.

A `Title` can added to figure with ```plt.title('...')```

````python

# create an array with 100 elements between 0 and 50
x = np.linspace(0, 50, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

plt.plot(x, y1)
plt.title("This is a title")
````

````python
# X and Y labels can be added to  figure by plt.xlabel('...') or plt.ylabel('..')


plt.plot(x, y2)
plt.xlabel('This is X label')
plt.ylabel('this is Y label')
````

`matplotlib.pyplot.legend()` is used in Python’s Matplotlib library to add a legend to a plot. A legend helps identify different plot elements by labeling them.

````python
plt.plot(x, y1, label = 'Y1')
plt.plot(x, y2, label = 'Y2')
plt.xlabel('This is a X Label')
plt.ylabel('This is a Y Label')
plt.legend(loc='upper right')
````

#### Line Colors and Styles

You can add the color to the line plots using `color` parameter, color codes(rgbcmyk), hex codes, and grayscale level.

```
Color codes (RGBCYMK):

'b'....blue
'g'....green
'r'....red
'c'....cyan
'm'....magenta
'y'....yellow
'k'....black
```
Hex codes:

Example: #00FF00' is green. Google any hex code color. 

````python
plt.figure(figsize=(10, 8))

plt.plot(x, y1, color='red') # red
plt.plot(x, y2, color='y')
plt.plot(x, y2*4, color='c') # cyan
plt.plot(x, y2*6, color='k') # black
````

#### Line styles and markers

```
Line styles

'-' ....solid line style
'--'....dashed line style
'-.'....dash-dot line style\
':'....dotted line style
'b'...# blue markers with default shape
'or'...# red circles
'-g'....# green solid line
'--'...# dashed line with default color
'^k:'...# black triangle_up markers connected by a dotted line
```

```
Line Markers


'.'...point marker
','...pixel marker
'o'...circle marker
'v'...triangle_down marker
'^'...triangle_up marker
'<'....triangle_left marker
'>'....triangle_right marker
'1'....tri_down marker
'2'....tri_up marker
'3'....tri_left marker
'4'....tri_right marker
's'....square marker
'p'....pentagon marker
'*'....star marker
'h'....hexagon1 marker
'H'....hexagon2 marker
'+'....plus marker
'x'....x marker
'D'....diamond marker
'd'....thin_diamond marker
'|'....vline marker

```



````python

# ls: Line Style
# lw: Line wdidth


plt.figure(figsize=(10, 8))


plt.plot(x, y1, color='red', ls='solid', marker='+', lw=3, markersize=3)
plt.plot(x, y2, color='b', ls='dotted', marker='D', lw=4, markersize=4) 
plt.plot(x, y2*2, color='#00FF00', ls='dashed', marker='*', lw=5, markersize=5) 
plt.plot(x, y2*3, color='0.8', ls=':', marker='s', lw=6, markersize=6)
````

#### Text and Annoation

If you want to add texts or annotation in visualization, Matplotlib can allow you to do that.

````python
### x = [1, 2, 3, 4, 5, 6, 7, 8]
y = np.square(x)

# plot x and y
plt.plot(x, y)


# To addd the text to the plot
# plt.plot(x position, y position, string, color or style)

plt.text(1.5, 55, 'This texts is added at the plot where x=1., y=30', color='r')
````
